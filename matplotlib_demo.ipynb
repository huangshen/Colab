```python
import matplotlib.pyplot as plt
import numpy as np

fs=50
t=np.arange(0,2,1.0/fs)
y=np.sin(2*np.pi*t)

fig,axe=plt.subplots()
axe.plot(t,y)
axe.set_title("matplotlib demo")
axe.set_xlabel("time(s)")
axe.set_ylabel("amplitude")
```
